```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('tidu.csv',index_col=None,converters={'时间':str})
oridata=pd.read_csv('Database.csv',index_col=None)
fig_list=['fig_%s' %i for i in range(oridata.shape[0])]
```

```python
j=0
i=0
while j<oridata.shape[0]:
    x=data.iloc[i:i+7,0].values
    y=data.iloc[i:i+7,1]
    fig,ax=plt.subplots()
    ax.step(x,y,linewidth=1)
    ax.set(xlim=(0,x[-1]),ylim=(0,100))
##水黑 乙醇红 乙腈黄 甲醇青 
    if oridata.iloc[j,54]=='乙醇':
        plt.fill_between(x,y,color='black',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='red',step='mid')
    elif oridata.iloc[j,54]=='乙腈':
        plt.fill_between(x,y,color='black',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='yellow',step='mid')
    elif oridata.iloc[j,54]=='甲醇':
        plt.fill_between(x,y,color='black',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='cyan',step='mid')
    elif oridata.iloc[j,51]=='乙醇':
        plt.fill_between(x,y,color='red',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='black',step='mid')
    elif oridata.iloc[j,51]=='乙腈':
        plt.fill_between(x,y,color='yellow',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='black',step='mid')
    elif oridata.iloc[j,51]=="甲醇":
        plt.fill_between(x,y,color='cyan',step='mid')
        plt.fill_between(x,y,plt.ylim()[1],color='black',step='mid')
    path='D:/OneDrive/SRTP/3模型/梯度图/%s.jpg'%(fig_list[j])
    plt.savefig(path)
    j+=1
    i+=11
```
