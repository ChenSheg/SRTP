```python
import joblib
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import PyPDF2
# 加载模型
model = joblib.load('XGBoost_model.joblib')
encoder=OneHotEncoder()

# 获取特征输入
中药编码=input("请输入中药名称：")


# 打开PDF文件
pdf_file = open('GBT 31774-2015.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file)

# 搜索PDF文件
search_text = 中药编码
search_results = []
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text = page.extract_text()
    if search_text in text:
        search_results.append((page_num+1))
# 输出搜索结果
if search_results:
    print('Search results for "{}":'.format(search_text))
    for result in search_results:
        print('Page {}'.format(result))
else:
    print('No results found for "{}"'.format(search_text))

# 关闭PDF文件
pdf_file.close()



```
