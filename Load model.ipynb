```python
import joblib
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# 获取特征输入
中药编码=input("请输入中药名称：")

# 读取Excel表格

medicine_df = pd.read_excel('GBT 31774-2015.xlsx', sheet_name='Sheet1')
pd.set_option('display.float',lambda x : '%.0f' % x)

# 提取特定特征值的行
medicine_filtered_df = medicine_df.query("品名 == '{}'".format(中药编码))

# 提取特定列的数据
medicine_column_select = medicine_filtered_df.iloc[:,3:11]
```

```python
# 假设你已经手动输入了一行未经处理的标签数据
new_data = pd.DataFrame(medicine_column_select)

# 使用原始数据集的列名来获取dummy变量
original_data = pd.read_csv('x_encode.csv')
dummy_data = pd.get_dummies(new_data.reindex(columns=original_data.columns),
                             columns=original_data.columns,dummy_na=False)

# 如果输入的标签有原数据库中不存在的，需要在dummy变量中添加缺失的列
missing_columns = set(original_data.columns) - set(dummy_data.columns)
for column in missing_columns:
    dummy_data[column] = 0



# 按照训练集的特征顺序来排列测试集的特征顺序
myinput = dummy_data.reindex(columns=original_data.columns)
myinput.drop('Unnamed: 0',axis=1,inplace=True)

# 加载模型
nucleoside_model = joblib.load('predictnucleoside.joblib')

# 进行预测
prediction = nucleoside_model.predict(myinput)
Label=pd.read_csv('Label.csv')

df = pd.DataFrame(prediction)

df.columns=pd.DataFrame(Label).columns

# 获取内容为1的列的列名
result = list(df.columns[df.eq(1).any()])
nucleoside=pd.DataFrame(result,columns=['核苷种类'])
```

```python
色谱柱名称=input("请输入色谱柱名称：")



# 读取Excel表格

column_df = pd.read_excel('column.xlsx', sheet_name='Sheet1')

# 提取特定特征值的行

filtered_df = column_df.query("Name == '{}'".format(色谱柱名称))

# 提取特定列的数据

column_select = filtered_df.iloc[:,-9:-3]
```

```python
##输入一大堆东西
A溶剂=input("请输入A溶剂名称：")
A添加剂种类mmol=input("请输入A中添加剂种类(以mmol为数量单位)：")
A添加剂种类percent=input("请输入A中添加剂种类（以含量计）：")
B溶剂=input("请输入B溶剂名称：")
B添加剂种类mmol=input("请输入B中添加剂种类(以mmol为数量单位)：")
B添加剂种类percent=input("请输入B中添加剂种类（以含量计）：")
needencodefeature=[A溶剂,A添加剂种类mmol,A添加剂种类percent,
                   B溶剂,B添加剂种类mmol,B添加剂种类percent]
encodedata=[needencodefeature]*nucleoside.shape[0]

# 将数据转换为 pandas 的 DataFrame 格式
df = pd.DataFrame(encodedata,columns=['A溶剂','A添加剂种类mmol','A添加剂种类percent',
                                      'B溶剂','B添加剂种类mmol','B添加剂种类percent'])

encode=pd.concat([nucleoside,df],axis=1)



# 使用原始数据集的列名来获取dummy变量
rongji_original_data = pd.read_csv('X_needencoder_dummy.csv')
rongji_dummy_data = pd.get_dummies(encode)

# 创建空的DataFrame
onehotresult = pd.DataFrame(columns=rongji_original_data.columns)

# 删除B中不存在于A中的列
for col in rongji_dummy_data.columns:
    if col not in rongji_original_data.columns:
        del rongji_dummy_data[col]

# 将B的数据填充到对应的列中
for col in rongji_dummy_data.columns:
    if col in onehotresult.columns:
        onehotresult[col] = rongji_dummy_data[col]

# 将缺失值填充为0
onehotresult.fillna(0, inplace=True)
```

```python
A添加剂种类浓度mmol=input("请输入A添加剂种类浓度mmol：")
Apercent=input("请输入A添加剂种类浓度%：")
B添加剂种类浓度mmol=input("请输入B添加剂种类浓度mmol：")
Bpercent=input("请输入B添加剂种类浓度%：")
柱温=input("请输入柱温：")
时间min=input("请输入时间min：")
比例A=input("请输入比例A：")
时间min1=input("请输入时间min.1：")
比例A1=input("请输入比例A.1：")
时间min2=input("请输入时间min.2：")
比例A2=input("请输入比例A.2：")
时间min3=input("请输入时间min.3：")
比例A3=input("请输入比例A.3：")
时间min4=input("请输入时间min.4：")
比例A4=input("请输入比例A.4：")
时间min5=input("请输入时间min.5：")
比例A5=input("请输入比例A.5：")
时间min6=input("请输入时间min.6：")
比例A6=input("请输入比例A.6：")
时间min7=input("请输入时间min.7：")
比例A7=input("请输入比例A.7：")
时间min8=input("请输入时间min.8：")
比例A8=input("请输入比例A.8：")
时间min9=input("请输入时间min.9：")
比例A9=input("请输入比例A.9：")
流速=input("请输入流速：")
进样量=input("请输入进样量：")

datafeature=[A添加剂种类浓度mmol,Apercent,B添加剂种类浓度mmol,Bpercent,
             柱温,时间min,比例A,时间min1,比例A1,时间min2,比例A2,时间min3,比例A3,
             时间min4,比例A4,时间min5,比例A5,时间min6,比例A6,时间min7,比例A7,
             时间min8,比例A8,时间min9,比例A9,流速,进样量]

inputdata=[datafeature]*nucleoside.shape[0]
column_select_data=[column_select]*nucleoside.shape[0]
column_select_data=pd.DataFrame(column_select_data,
                                columns=['H','S*','A','B','C (pH 2.8)','C (pH 7.0)'])


inputdata=pd.DataFrame(inputdata,columns=['柱温','A添加剂种类浓度mmol',
                       'A添加剂种类浓度percent','B添加剂种类浓度mmol',
                       'B添加剂种类浓度percent','时间min','比例A','时间min1',
                       '比例A1','时间min2','比例A2','时间min3','比例A3','时间min4',
                       '比例A4','时间min5','比例A5','时间min6','比例A6',
                       '时间min7','比例A7','时间min8','比例A8',
                        '时间min9','比例A9','流速','进样量'])


# 合并特征和标签特征

features = pd.concat([onehotresult,inputdata,column_select_data], axis=1)


# 保存特征数据
features.to_excel('features.xlsx')
```

```python
# 加载模型
model = joblib.load('XGBoost_model.joblib')
```

```python
X=pd.read_excel('X.xlsx')

# 删除B中重复的标签
features = features.loc[:, ~features.columns.duplicated()]


featureinput=pd.DataFrame(columns=X.columns)

features=features.reindex(columns=X.columns,fill_value=0)

for col in features.columns:
    if col in featureinput.columns:
        featureinput[col]=features[col]

# 进行预测
prediction = model.predict(featureinput)

# 输出预测结果
print('预测结果为：', prediction)
```
